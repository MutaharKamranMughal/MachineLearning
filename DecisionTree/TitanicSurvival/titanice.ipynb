```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
new_df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
new_df.head(20)
```

```python
new_df.isnull().sum()
```

```python
mean_age = math.floor(df['Age'].mean())
new_df['Age'].fillna(mean_age, inplace=True)
new_df.head(10)
```

```python
new_df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.linear_model import LogisticRegression as lr
```

```python
le_sex = LabelEncoder()
new_df['Sex_train'] = le_sex.fit_transform(new_df['Sex'])
new_df
```

```python
final_df = new_df.drop('Sex', axis='columns')
final_df
```

```python
X = final_df[['Pclass', 'Age', 'Fare', 'Sex_train']]
X.head()
```

```python
y = final_df.Survived
y.head()
```

```python
#model = DecisionTreeClassifier()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#model.fit(X_train, y_train)
```

Determining the Best Model:

```python
def best_model(x_train, y_train, x_test, y_test):
    rfc_classifier = rfc()
    logistic_nigga = lr()
    decisive_nigga = dtc()
    
    rfc_classifier.fit(x_train, y_train)
    logistic_nigga.fit(x_train, y_train)
    decisive_nigga.fit(x_train, y_train)
    
    dickshionary = {
        'Decision Tree' : rfc_classifier.score(x_test, y_test),
        'logistic regression' : logistic_nigga.score(x_test, y_test),
        'decision tree' : decisive_nigga.score(x_test, y_test)
    } 
    
    high_lst = []
    model_name = []
    for model, accuracy in dickshionary.items():
        high_lst.append(accuracy)
        model_name.append(model)
    print(f'best model: {model_name[np.argmax(high_lst)], np.max(high_lst)}')
```

```python
best_model(X_train, y_train, X_test, y_test)
```
