```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
# Making dummy variable column:

dummies = pd.get_dummies(df.town) # to create dummies.
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')  # to join old and new data frames
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
X = final.drop('price', axis='columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800, 0 , 1]])
```

```python
model.predict([[3400, 0, 0]])
```

```python
model.score(X,y) * 100  # to check the accuracy of the model.
```
